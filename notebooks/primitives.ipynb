```python
from sudoku import Sudoku
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Create a sudoku

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
plt.subplot(121)
Sudoku().draw()
plt.title('Empty sudoku')
plt.subplot(122)
sudoku = Sudoku.random()
sudoku.draw()
plt.title('Random sudoku');
```

##  Sample sudoku cells

We are using the sudoku defined above (the second one)

```python
print("Values on the first row:")
for k in range(0, 9):
    cell = sudoku[0, k]
    print(cell, end='\t' if (k+1) % 3 != 0 else '\t\t')
print()
```

## Modify sudoku cells 

Set all numbers in the first column to 1, 2, 3, ..., 9

```python
sudoku[:, 0] = np.arange(1, 10)
sudoku.show(figsize=(4, 4))
```

## Delete sudoku cells

Remove all numbers in the middle square

```python
del sudoku.squares[1, 1]
sudoku.show(figsize=(4, 4))
```

You can do the same to remove single cells, rows or columns

```python
del sudoku[0, 0]
del sudoku.cols[-1]
del sudoku[:, -1] # Same the line above
del sudoku.rows[-1]
del sudoku[-1] # Same as the line above
sudoku.show(figsize=(4, 4))
```

## Iterating through cells

Iterate over all cells in the sudoku (first it iterates over rows then columns)

```python
', '.join([str(cell) for cell in sudoku])
```

Iterate over each row

```python
for i, row in enumerate(sudoku.rows):
    print('Row {}:  '.format(i) + '  '.join([str(cell) for cell in row]))
```

Iterate over each column

```python
for j, column in enumerate(sudoku.columns):
    print('Column {}:  '.format(j) + '  '.join([str(cell) for cell in column]))
```

Iterate over each square (squares are 3x3 arrays but here we display them flattened)

```python
for k, square in enumerate(sudoku.squares):
    print('Square {}:  '.format(k) + '  '.join([str(cell) for cell in square.flatten()]))
```

More complex operations can be performed. e.g: Get all numbers in the middle of each square

```python
print(sudoku[1::3, 1::3])
```
