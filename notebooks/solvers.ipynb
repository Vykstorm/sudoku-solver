```python
import numpy as np
import matplotlib.pyplot as plt
from sudoku import Sudoku
from dataset import SudokuDataset
from solvers import BasicSudokuIterativeSolver
from IPython.display import HTML, display

%matplotlib inline
```

## Solve a sudoku

```python
unsolved, solved = SudokuDataset().get_sample(return_solution=True)
```

```python
plt.subplots(1, 2, figsize=(8, 4))
plt.subplot(121)
unsolved.draw()
plt.subplot(122)
solved.draw();
```

Lets solve the sudoku with our baseline algorithm...

```python
solver = BasicSudokuIterativeSolver()
sudoku = unsolved.copy()
solver.solve(sudoku)
```

Now we compare it again with the real solution

```python
plt.subplots(1, 2, figsize=(8, 4))
plt.subplot(121)
sudoku.draw()
plt.subplot(122)
solved.draw();
```
