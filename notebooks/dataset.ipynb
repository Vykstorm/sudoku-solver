```python
from sudoku import Sudoku
from dataset import SudokuDataset
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

%matplotlib inline
```

## Import sudoku from file

```python
sudoku = Sudoku.fromfile('quizz.txt')
sudoku.show(figsize=(5, 5))
```

## Get sudokus from dataset

Get valid unsolved sudoku configurations

```python
samples = SudokuDataset().get_samples(return_solutions=False, shuffle=True)

plt.subplots(2, 4, figsize=(16, 8));
for k in range(0, 8):
    plt.subplot(2, 4, k+1)
    sudoku = next(samples)
    sudoku.draw()
```

Get can request also the solutions for each sudoku

```python
samples = SudokuDataset().get_samples(return_solutions=True, shuffle=True)

plt.subplots(3, 2, figsize=(8, 12));
for k in range(0, 3):
    sudoku, solution = next(samples)
    plt.subplot(3, 2, k*2+1)
    sudoku.draw()
    plt.subplot(3, 2, (k+1)*2)
    solution.draw()
```
